# Protype code to validate the Kmeans Clustering Algorithm
We will use the Yahoo! dataset to validate the Kmeans Clustering Algorithm, and we will use the following steps:
- Load the Yahoo! dataset
- Implement the Kmeans Clustering Algorithm
- Validate the Kmeans Clustering Algorithm
- Analyze the results
- Try various values of K

## Number of Rows = 1000

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load dataset
file_path = "./input/ydata-ysm-keyphrase-bid-imp-click-v1_0"
df = pd.read_csv(
    file_path,
    delimiter="\t",
    header=None,
    names=[
        "day",
        "account_id",
        "rank",
        "keyphrase",
        "avg_bid",
        "impressions",
        "clicks",
    ],
    nrows=1000,
)

# Assuming 'avg_bid', 'impressions', 'clicks' are the features we're interested in
X = df[["avg_bid", "impressions", "clicks"]].values

# Choose a range of cluster sizes to evaluate
ks = range(1, 10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(X)
    inertias.append(model.inertia_)

# Plot the elbow graph to find the optimal k
plt.figure(figsize=(8, 4))
plt.plot(ks, inertias, "-o")
plt.xlabel("Number of clusters, k")
plt.ylabel("Inertia")
plt.xticks(ks)
plt.title("Elbow Method For Optimal k")
plt.show()
```

```python
import plotly.express as px
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import plotly.io as pio
pio.renderers.default = 'notebook'

# Assume X is already defined and contains the data.
k_optimal = 3  # Replace with the actual optimal k

# Apply log transformation with an offset to handle zero values
offset = 1
X_log = np.log(X + offset)

# Fit the KMeans model on the log-transformed data
model_optimal_log = KMeans(n_clusters=k_optimal)
model_optimal_log.fit(X_log)
y_kmeans_log = model_optimal_log.predict(X_log)

# Create a DataFrame for Plotly with log-transformed data and clusters
df_plot = pd.DataFrame(X_log, columns=['Log Avg Bid', 'Log Impressions', 'Log Clicks'])
df_plot['Cluster'] = y_kmeans_log

# Create an interactive 3D scatter plot with Plotly
fig = px.scatter_3d(df_plot, x='Log Avg Bid', y='Log Impressions', z='Log Clicks',
                    color='Cluster', symbol='Cluster',
                    title='3D Clusters Visualization with Log Transformation')

# Customize the marker size
fig.update_traces(marker=dict(size=3))

# Calculate the log-transformed centroids
centroids_log = np.log(model_optimal_log.cluster_centers_ + offset)

# Add centroids to the plot
for idx, centroid in enumerate(centroids_log):
    fig.add_trace(
        px.scatter_3d(pd.DataFrame([centroid], columns=['Log Avg Bid', 'Log Impressions', 'Log Clicks']),
                      x='Log Avg Bid', y='Log Impressions', z='Log Clicks').update_traces(marker=dict(size=10, color='red'), showlegend=True).data[0]
    )

# Show the plot
fig.show()
```
